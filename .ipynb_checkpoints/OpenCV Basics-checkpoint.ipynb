```python
img = cv2.imread("squid.jpg")
```

```python
plt.imshow(img)
```

```python
temp = img.copy()
temp = cv2.flip(temp, 0) # Horizontal flip
plt.imshow(temp)
```

```python
temp = img.copy()
temp = cv2.flip(temp, 1) # Vertical flip
plt.imshow(temp)
```

```python
temp = img.copy()
temp = cv2.flip(temp, -1) # Horizontal and Vertical flip
plt.imshow(temp)
```

```python
img_black = np.zeros((100, 100, 3), dtype = np.int32)
```

```python
plt.imshow(img_black)
```

```python
temp = img_black.copy()
cv2.circle(
    temp,
    center = (40, 30),
    radius = 10,
    color = (255, 0, 0),
    thickness = 2
)
plt.imshow(temp)
```

```python
temp = img_black.copy()
cv2.rectangle(
    temp,
    pt1 = (40, 40),  
    pt2 = (30, 30),
    color = (0, 255, 0),
    thickness = 2
)
plt.imshow(temp)
```

```python
vertices = np.array([
    [10, 10],
    [20, 40],
    [60, 30]
])

pts = vertices.reshape(-1, 1, 2) # Reshape as necessary

temp = img_black.copy()
cv2.polylines(
    temp,
    [pts],
    isClosed = True,
    color = (0, 0, 255),
    thickness = 3
)
plt.imshow(temp)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
temp = img_black.copy()
cv2.putText(
    temp,
    text = "Akshit",
    org = (5, 80),
    fontFace = font,
    fontScale = 1,
    lineType = cv2.LINE_AA,
    color = (0, 255, 0)
)
plt.imshow(temp)
```
